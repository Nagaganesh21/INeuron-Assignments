```python
class Triangle:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
        
        
class Tr_Area(Triangle):
    def area(self):
        s = (self.a+self.b+self.c)/3
        return (s*(s-self.a)*(s-self.b)*(s-self.c))**0.5
  
```

```python
a = int(input('Length of side 1:'))
b = int(input('Length of side 2:'))
c = int(input('Length of side 3:'))
Area = Tr_Area(a,b,c)
print('The area of the Triangle is:{0}'.format(Area.area()))
```

```python
def filter_long_words(list,n):
    l = []
    for i in list:
        if len(i)>n:
            l.append(i)
    return l

print(filter_long_words(['Sriram','AP','ESV'],3))
```

```python
print(filter_long_words(['Hyderabad','Chennai','Bangalore','Delhi','Vizag','Pune'],5))
```

```python
def word_map(list):
    l = []
    for i in list:
        l.append(len(i))
    return l
print(word_map(['Sriram','Esv']))    
```

```python
print(word_map(['Deva']))
```

```python
def vow(n):
    if len(n)>1:
        return False
    else:
        if n in 'aeiou':
            return True
        else:
            return False
print(vow('a'))
```
