```python
def div(a,b):
    try:
        res = a/b
        print(res)
    except ZeroDivisionError:
        print("Sorry ! Can't divide by 0")
        
```

```python
#Declaration
subjects = ['Americans','Indians']
verbs = ['play','watch']
objects = ['Baseball','Cricket']

comb = len(subjects)*len(verbs)*len(objects)
```

```python
#Defining a function
def bin_enc(n):
    l = []
    b = bin(n).replace('0b','')
    if len(b) == 1:
        l.append(0)
        l.append(0)
        l.append(int(b))
    elif len(b) == 2:
        l.append(0)
        l.append(int(b[0]))
        l.append(int(b[1]))
    else:
        l.append(int(b[0]))
        l.append(int(b[1]))
        l.append(int(b[2]))
    return l
        
```

```python
#Test
bin_enc(7)
```

```python
#Final execution
for i in range(comb):
    b = bin_enc(i)
    print(subjects[b[0]]+' '+verbs[b[1]]+' '+objects[b[2]])
```
